```python
import numpy as np
import pandas as pd
from scipy.stats import norm
```

```python
model_experiment = pd.read_csv(r'/Users/meghanchilds/Desktop/Dissertation_Research/COVID-19-Model-Formal-Calibration/MCMC CSV Files/Model_experiment_data_frame_rerun_APR_24.csv', names =['days', 'tests administered', 'positive test results', 'isolation pool', 'models proba pos test'])

model_experiment
```

```python
model_experiment = model_experiment.drop(labels = np.nan, axis =0)
model_experiment
```

```python
residuals = np.zeros(len(model_experiment.iloc[:,0]))

for i in range(len(model_experiment.iloc[:,0])):
    std = 2*float(model_experiment.iloc[i,1])*float(model_experiment.iloc[i,4])*(1-float(model_experiment.iloc[i,4]))
    residuals[i] = norm.rvs(loc = 0, scale = std, size =1)
```

```python
residuals
```

```python
np.round(residuals)
```

```python
model_experiment['positive test results'] = pd.to_numeric(model_experiment['positive test results']) + np.round(residuals)
```

```python
pd.to_numeric(model_experiment['positive test results'])
```
